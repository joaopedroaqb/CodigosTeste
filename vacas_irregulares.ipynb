```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
import torch
torch.__version__
import torch.nn as nn

# Set random seed for reproducibility
np.random.seed(123)
torch.manual_seed(123)

# Load data
healthycows = pd.read_csv('/path/to/healthycows.csv')
lumpycows = pd.read_csv('/path/to/lumpycows.csv')

# Combine the datasets
all_cows = pd.concat([healthycows, lumpycows])

# Separate features and labels
features = all_cows.drop('label', axis=1)
labels = all_cows['label']

# Split data into training and testing sets
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.25, random_state=123)

# Convert data to tensors
features_train = torch.tensor(features_train.values, dtype=torch.float)
labels_train = torch.tensor(labels_train.values, dtype=torch.float)
features_test = torch.tensor(features_test.values, dtype=torch.float)

# Define the dataset and data loaders
train_dataset = torch.utils.data.TensorDataset(features_train, labels_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=10, shuffle=True)

# Define the model architecture
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        self.fc1 = nn.Linear(30, 16)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(16, 16)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x

# Create an instance of the classifier
classifier = Classifier()

# Define the loss function and optimizer
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(classifier.parameters(), lr=0.001, weight_decay=0.0001)

# Training loop
for epoch in range(100):
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = classifier(inputs)
        loss = criterion(outputs, labels.unsqueeze(1))
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f'Epoch {epoch+1}: Loss {running_loss/len(train_loader):.5f}')

# Evaluation
classifier.eval()
with torch.no_grad():
    predictions = classifier(features_test)
    predictions = np.array(predictions > 0.5, dtype=int)

accuracy = accuracy_score(labels_test, predictions)
confusion_mat = confusion_matrix(labels_test, predictions)

print(f'Accuracy: {accuracy:.4f}')
print('Confusion Matrix:')
print(confusion_mat)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
import torch
torch.manual_seed(123)
import torch.nn as nn

# Load the dataset
np.random.seed(123)
previsores = pd.read_csv('/content/entradas_breast.csv')
classe = pd.read_csv('/content/saidas_breast.csv')

# Split the data into training and testing sets
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(
    previsores, classe, test_size=0.25, random_state=123)

# Convert the data to tensors
previsores_treinamento = torch.tensor(np.array(previsores_treinamento), dtype=torch.float)
classe_treinamento = torch.tensor(np.array(classe_treinamento), dtype=torch.float)

# Define the dataset and data loaders
dataset = torch.utils.data.TensorDataset(previsores_treinamento, classe_treinamento)
train_loader = torch.utils.data.DataLoader(dataset, batch_size=10, shuffle=True)

# Define the model architecture
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        self.fc1 = nn.Linear(in_features=30, out_features=16)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(16, 16)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x

# Create an instance of the classifier
classificador = Classifier()

# Define the loss function and optimizer
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(classificador.parameters(), lr=0.001, weight_decay=0.0001)

# Training loop
for epoch in range(100):
    running_loss = 0.
    for data in train_loader:
        inputs, labels = data
        optimizer.zero_grad()
        outputs = classificador(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f'Epoch {epoch+1}: Loss {running_loss/len(train_loader):.5f}')

# Save the PyTorch model
torch.save(classificador.state_dict(), 'pytorch_Model.pth')

# Load the saved PyTorch model
classificador = Classifier()
classificador.load_state_dict(torch.load('pytorch_Model.pth'))
classificador.eval()

# Load the Keras model
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np

model = load_model("keras_Model.h5", compile=False)

# Load the labels
class_names = open("labels.txt", "r").readlines()

# Load and preprocess the image
image_path = "<IMAGE_PATH>"
image = Image.open(image_path).convert("RGB")
size = (224, 224)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)
image_array = np.asarray(image)
normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
data[0] = normalized_image_array

# Predict using the Keras model
prediction = model.predict(data)
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]

# Print the Keras model's prediction and confidence score
print("Keras Model Prediction:")
print("Class:", class_name[2:])
print("Confidence Score:", confidence_score)

# Convert the image to a PyTorch tensor
image_tensor = torch.tensor(normalized_image_array, dtype=torch.float)
image_tensor = image_tensor.permute(2, 0, 1).unsqueeze(0)

# Predict using the PyTorch model
with torch.no_grad():
    output = classificador(image_tensor)
    pytorch_prediction = torch.sigmoid(output).item()

# Print the PyTorch model's prediction
print("PyTorch Model Prediction:")
print("Class: Healthy" if pytorch_prediction < 0.5 else "Class: Lumpy")
print("Confidence Score:", pytorch_prediction)
```
